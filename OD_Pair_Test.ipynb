```python
import numpy as np
import pandas as pd
from shapely.geometry import Polygon,box,Point
import os
os.environ['USE_PYGEOS'] = '0'
import geopandas as gpd
from shapely.wkt import loads

import osmnx as ox
import networkx as nx
from osmnx.distance import nearest_nodes

import matplotlib.pyplot as plt
import random
```

```python
# Load the data
# "C:\Users\zliu33\Downloads\OD_pairs.csv"
file_path = r'C:\Users\zliu33\Downloads\OD_pairs_new.csv'
od_pairs_new = pd.read_csv(file_path)
```

```python
od_pairs_new.head()
```

```python
# Assuming the loaded data already has Point objects, create the GeoDataFrame directly
gdf_new = gpd.GeoDataFrame(od_pairs_new, geometry='destination', crs='EPSG:26986')

# Transform destination to EPSG:4326 for geographic coordinates
gdf_new = gdf_new.to_crs(epsg=4326)
gdf_new['Destination_Lon'] = gdf_new.geometry.x
gdf_new['Destination_Lat'] = gdf_new.geometry.y

# Set and transform the geometry for the centroid
gdf_new = gdf_new.set_geometry('centroid', crs='EPSG:26986')
gdf_new = gdf_new.to_crs(epsg=4326)
gdf_new['Centroid_Lon'] = gdf_new.geometry.x
gdf_new['Centroid_Lat'] = gdf_new.geometry.y

# Display the DataFrame with the new longitude and latitude columns
gdf_new.head()
```

```python
# Export OD pairs to CSV
od_pairs_path = r'C:\Users\zliu33\Downloads\OD_lonlat.csv'
gdf_new.to_csv(od_pairs_path, index=False)
```
